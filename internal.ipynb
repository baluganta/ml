Question 1

get android malvare detection  data set from caggle and apply pre processing and feature extraction   (pca) with no of extracted components  100 then construct a classification  model using logestic regression  and measure its performance using varios performance measures using python language

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('/content/Android Malware.csv')

print(df.isna().sum())
df.dropna(inplace=True)
df.replace({'malware': 0, 'goodware': 1}, inplace=True)
x,y = df.drop(columns=['Label'], axis=1),df['Label']
scaler = StandardScaler()
x_change = scaler.fit_transform(x)
pca = PCA(n_components=100)
x_pca = pca.fit_transform(x_change)
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.30)
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred,average='weighted'))
print(f1_score(y_test,y_pred,average='weighted'))
print(recall_score(y_test,y_pred,average='weighted'))
print(confusion_matrix(y_test,y_pred))
```

Get cirrhosis patient survival prediction dataset from kaggle and apply preprocessing and feature selection with number of features 15 and train the classification by using naive bayes classifier and measure it's performance using various performance measures
Note:- Tune the parameters of the model to get maximum performance

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.feature_selection import RFE
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd

```

```python
df = pd.read_csv('/content/Cirrhosis Data.csv')
print(df.columns)
#print(df.isna().sum())
#df
columns_to_drop = ['encounter_id', 'patient_id', 'hospital_id', 'icu_id', 'Unnamed: 83']
df = df.drop(columns=columns_to_drop)
df.isnull().sum()
```

```python
df.head()
```

```python
for column in df.columns:
    if df[column].dtype == 'float64':
        # Handle numeric columns (replace NaN with mean)
        df[column].fillna(df[column].mean(), inplace=True)
    elif df[column].dtype == 'object':
        # Handle categorical columns (replace NaN with mode)
        df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df["hospital_death"].value_counts()
```

```python
#df['Platelets'] = df['Platelets'].fillna(df['Platelets'].mean())
#df['Prothrombin'] = df['Prothrombin'].fillna(df['Prothrombin'].mean())
#df['Stage'] = df['Stage'].fillna(df['Stage'].mode()[0])

df.dropna(inplace=True)
print(df.isna().sum())
x,y = df.drop('hospital_death',axis=1),df['hospital_death']
oe = preprocessing.OrdinalEncoder()
x_change = oe.fit_transform(x)
scaler = preprocessing.StandardScaler()
x_change = scaler.fit_transform(x_change)
print(x_change)
```

```python
es = LogisticRegression(solver='lbfgs', multi_class='auto',max_iter=1000)
select = RFE(estimator=es,n_features_to_select=15)
features = select.fit(x_change,y)
features = x.columns[features.support_]
#print(features)
x = x[features]
oe = preprocessing.OrdinalEncoder()
x = oe.fit_transform(x)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30)
model = GaussianNB(var_smoothing=1e-2)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred,average='weighted'))
print(f1_score(y_test,y_pred,average='weighted'))
print(recall_score(y_test,y_pred,average='weighted'))
print(confusion_matrix(y_test,y_pred))
```

Question 3

Get animal condition classification dateset from Kaggle and perform preprocessing and train 5 classification models using of different ML algorithms (NA, LR, SVM, OT, KNN with K=7) and join them using majority voting and find out the performance of that ensemble learner.

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd
```

```python
df = pd.read_csv('/content/Animaldata.csv')
print(df.columns)
print(df.isna().sum())
df.dropna(inplace=True)
x,y = df.drop('Dangerous',axis=1),df['Dangerous']
oe = preprocessing.OrdinalEncoder()
x = oe.fit_transform(x)
y.replace({'Yes':0,'No':1},inplace=True)
```

```python
models=[]
pred = []
models.append(LogisticRegression())
models.append(DecisionTreeClassifier())
models.append(RandomForestClassifier(n_estimators=4))
models.append(KNeighborsClassifier(n_neighbors=7))
models.append(GaussianNB())
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30)
for i in models:
  i.fit(x_train,y_train)
  pred.append(i.predict(x_test))
#print(models)
#print(pred)
y_pred = 0
for i in pred:
  y_pred += i
print(y_pred)
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred,average='weighted'))
print(f1_score(y_test,y_pred,average='weighted'))
print(recall_score(y_test,y_pred,average='weighted'))
print(confusion_matrix(y_test,y_pred))
```
